# Choosing representative sample set

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import requests
import time
import random
import seaborn
%matplotlib notebook
# Import API key
from config import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)



# Range of latitudes and longitudes
```

```python
world = pd.read_csv("resources/worldcities.csv")
world.head()
```

```python
cities = pd.read_csv("resources/cities.csv")
cities.head()
```

```python
randomCities = world.sample(n=500)
randomCities.head()
```

```python
#Choosing the weather outcomes for the cities

rowCounter = 0

randomCities["Temperature"] = ""
randomCities["Humidity"] = ""
randomCities["Cloudiness"] = ""
randomCities["Windspeed"] = ""

for index, row in randomCities.iterrows():
    time.sleep(1)
    base_url = "http://api.openweathermap.org/data/2.5/weather?"
    units = "Imperial"
    query_url = base_url + "lat=" + str(row["Latitude"]) + "&lon=" + str(row["Longitude"]) + "&appid=" + api_key + "&units=" + units
    
    print("Count" + str(rowCounter))
    print(query_url)
    rowCounter+= 1
    weather = requests.get(query_url).json()
    
    try:
        temperature = weather["main"]["temp"]
        humidity = weather["main"]["humidity"]
        cloudiness = weather["clouds"]["all"]
        wind_speed = weather["wind"]["speed"]
        
        randomCities.set_value(index, "Temperature", temperature)
        randomCities.set_value(index,"Humidity", humidity)
        randomCities.set_value(index,"Cloudiness", cloudiness)
        randomCities.set_value(index,"Wind Speed", wind_speed)
    except:
        print("Skipped weather.")
randomCities
                                
```

```python
randomCities.to_csv("AroundTheWorld.csv", index = False)
```

```python
plt.scatter(randomCities["Latitude"], randomCities["Temperature"], alpha = 0.5)

plt.title("Temperature (F) vs. Latitude Date of Analysis 10-29-2017")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.savefig("Temperature.png")
plt.show
```

```python
plt.scatter(randomCities["Latitude"],randomCities["Wind Speed"], alpha= 0.75)
plt.title("Latitude vs Wind Speed Date of Analysis 10-29-2017")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed")
plt.savefig("Wind Speed.png")
plt.show()
```
